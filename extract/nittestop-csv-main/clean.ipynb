```python
df1["from"]="Karkala"
df1 = df1.drop(["time"],axis=1)
df1.head()
```

```python
df2 = pd.read_csv("From Nitte to Belman-Padubidri (Udupi-Mangalore) 2c4fffeaf3dd4f54a91b2d2a9e04ca7e.csv")
df2.head()
```

```python
df2.columns = ["at_nitte","name","type","destination"]
df2["from"]="Karkala"
df2.head()
```

```python
df3 = pd.read_csv("From Nitte to Karkala b05bfb8455d249c5a9cf73d278fd9cfd.csv")
df3.head()
```

```python
df3.columns = ["at_nitte","name","type","from"]
df3["destination"]="Karkala"
df3.head()
```

```python
final_df = pd.concat([df1,df2,df3],axis=0)
```

```python
def parseTime(x):
    h,m = x.split(":")
    return [int(h),int(m)]

time_mat = list(map(parseTime,final_df.at_nitte))
```

```python
def to_24(x):
    h,m = x.split(":")
    h,m = int(h),int(m)
    if(h<6):
        h=h+12
    return "{hour:02d}:{min:02d}".format(hour=h,min=m)

final_df.at_nitte=final_df.at_nitte.apply(to_24)
```

```python
final_df.head()
```

```python
sorted_df = final_df.sort_values(by="at_nitte")
sorted_df.head()
```

```python
def to_24(x):
    h,m = x.split(":")
    h,m = int(h),int(m)
    if(h>12):
        h=h-12
    return "{hour:02d}:{min:02d}".format(hour=h,min=m)

sorted_df.at_nitte=sorted_df.at_nitte.apply(to_24)

sorted_df.head()
```
